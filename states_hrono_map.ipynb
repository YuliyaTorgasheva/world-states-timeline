```python
import pandas as pd
#словарь с соответствием - прекращение гос в краткой форме
v1 = pd.read_excel('Словарь соответствий.xlsx')
v1
```

```python
v_full = pd.read_excel('Эрлихман от Никиты Максимова V2.xlsx')
v_full['Прекращение государственности'] = v_full['Прекращение государственности'].str.replace('.','')
v_full['Прекращение государственности'] = v_full['Прекращение государственности'].str.replace('ё','е')
v_full['Прекращение государственности'] = v_full['Прекращение государственности'].str.replace('Присоединение','Присоединено')
v_full
```

```python
#к формудировке прекращения (присодеинения или др) добавляем краткое название преемника
#НВ - убрать (наше время)
final = v_full.merge(v1,left_on='Прекращение государственности',right_on='Статус и зависимость',how='left')
print(final)
final = final[['Государство','target','Дата начала','Дата окончания','Регион']]
final = final.dropna()
final = final.rename(columns={
    'Государство': 'source',
    'target': 'target',
    'Дата начала': 'start',
    'Дата окончания': 'end',
    'Регион': 'region'
})
final = final[pd.to_numeric(final['start'], errors='coerce').notnull()]
final['start'] = pd.to_numeric(final['start'])
# final['end'] = final['end'].str.strip().str.lower()
final['end'] = final['end'].replace({'нв': 2025, 'наше время': 2025})
# final['end'] = pd.to_numeric(final['end'], errors='coerce')
final['target'] = final['target'].replace({2025: None, 0: None})
final = final[~(final['end'] < final['start'])]
final = final.sort_values(by=['region','start', 'target'], ascending=[True, True, True])
final.to_csv('states_V7.csv', index=False)
final
```
